```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
data = pd.read_csv('event.csv', header=0, index_col=0, parse_dates=True, squeeze=False)
data.head(15)
```

```python
# 建立fig:畫布面板跟 axes:圖表
# Hint : subplot(2,1) 建立兩張圖表
fig, axes = plt.subplots(2, 1, figsize=(10,10))

# 圖表外觀設定 : seaborn-darkgrid
plt.style.use('seaborn-darkgrid')

# Hint axes[0]:第一張圖表；axes[1]:第二張圖表
axes[0].plot(data.index, data['temperature'], label = 'temperature', linewidth = 2, color='red')
axes[1].plot(data.index, data['pressure'], label = 'pressure', linewidth = 2, color='green')

# X軸旋轉字體
# Hint axes[0].set_xticklabels(index, rotation)
axes[0].set_xticklabels(data.index, rotation=45)
axes[1].set_xticklabels(data.index, rotation=45)

# WeekdayLocator(星期), DayLocator(日) 
# Hint axes[0].xaxis.set_major_locator(...)
axes[0].xaxis.set_major_locator(mdates.WeekdayLocator(interval=1))
axes[1].xaxis.set_major_locator(mdates.WeekdayLocator(interval=1))

# 設定時間格式
# Hint axes[0].xaxis.set_major_formatter(...)
axes[0].xaxis.set_major_formatter(mdates.DateFormatter('%Y/%m/%d'))
axes[1].xaxis.set_major_formatter(mdates.DateFormatter('%Y/%m/%d'))

# Hint :
# 設定圖表的標題 : axes[0].set_title
# 設定圖表x軸單位 : axes[0].set_xlabel
# 設定圖表y軸單位 : axes[0].set_ylabel
# 設定圖表圖示說明(位置及文字大小) : axes[0].legend
axes[0].set_xlabel("Time(m)")
axes[0].set_ylabel("Value")
axes[0].set_title("Temperature")
axes[0].legend(loc=8)
axes[1].set_xlabel("Time(m)")
axes[1].set_ylabel("Value")
axes[1].set_title("Pressure")
axes[1].legend(loc=8)


# 儲存成圖片，dpi為解析度
plt.savefig("Temperature_Pressure.png", dpi = 500)
```
