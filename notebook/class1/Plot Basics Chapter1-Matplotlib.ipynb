## 大綱:資料視覺化

### 三大套件:Matplotlib、Seaborn、Plotly

* Matplotlib
    * 優點 : 畫圖功能齊全，網路教學豐富
    * 缺點 : 程式架構複雜(美化工程)

* Seaborn (在 matplotlib 的基礎上進行更高級的 API 封裝)
    * 優點 : 程式建構簡單，圖表美化容易
    * 缺點 : 網路教學討論程度偏少，初學者不易學習

* Plotly
    * 優點 : 互動化的介面
    * 缺點 : 吃電腦資源
    
### 圖表分類 應用類型

* 連續變項的統計圖 : 表示具連續性相關的數據
    * 折線圖 (plot)
    * 直方圖 (hist)

* 觀察欄位之間關聯的性質
    * 散狀圖(scatter) : 也有分群密集的性質
    
* 類別變項的統計圖 : 表示與種類相關無連續性質的數據
    * 長條圖 (bar)
    * 圓餅圖 (pie)

* **盒鬚圖(boxplot)** : 可表示連續性相關的數據資訊及欄位之間的關聯
* **3D gif圖** : 以3D動態呈現

### 線條 style

* 顏色 : `color` 例: red : `r`, blue : `b`, green : `g`
* 粗細 : `linewidth` 預設為1
* 線的格式 : `linestyle` 例: 實線 : `solid`, 虛線 : `dotted`, 其他格式 : `-.`
* 點狀形式 : `marker` 例: 菱形 : `D`, 三角點 : `^`

```python
# 設定數據 在0 <= x < 2 範圍中間距為 0.4 : [0., 0.4, 0.8, 1.2, 1.6]
t = np.arange(0., 2., 0.4)

# x : t ; y : t**1 ; color : red ; linewidth : 0.9 ; linestyle : 實線
plt.plot(t, t**1, color='r', linewidth=0.9, linestyle='solid')

# x : t ; y : t**2 ; color : blue ; linewidth : 2.5 ; linestyle : 虛線 ; marker : 菱形
plt.plot(t, t**2, color='b', linewidth=2.5, linestyle='dotted', marker='D')

# x : t ; y : t**3 ; color : green ; linewidth : 2 ; linestyle : -. ; marker : 三角點
plt.plot(t, t**3, color='g', linewidth=2, linestyle='-.', marker='^')
```

#### 可以快速取代多條 marker、line、color 的寫法

`plt.plot(x1, y1, '[marker1][line1][color1]', x2, y2, '[marker2][line2][color2]')`

```python
# 設定數據 在0 <= x < 2 範圍中間距為 0.4 : [0., 0.4, 0.8, 1.2, 1.6]
t = np.arange(0., 2., 0.4)

# '-r' : 實線紅色 ; 'D:b' : 菱形虛線藍色 ; '^-.g' : 三角點 -. 綠色
# x1 : t, y1 : t**1
# x2 : t, y2 : t**2
# x3 : t, y3 : t**3
plt.plot(t, t**1, '-r', t, t**2, 'D:b', t, t**3, '^-.g', linewidth=2)
```

### 新增圖形資訊

* 線的名稱 : `label`
* 圖的標題 : `plt.title`
* x軸單位 : `plt.xlabel`
* y軸單位 : `plt.ytable`
* 圖示說明(位置及文字大小) : `plt.legend`

```python
# 設定數據 在0 <= x < 2 範圍中間距為 0.4 : [0., 0.4, 0.8, 1.2, 1.6]
t = np.arange(0., 2., 0.4)
# 新增每條線的名稱 : label
plt.plot(t, t**1, color='r', linewidth=0.9, linestyle='solid', label='y1')
plt.plot(t, t**2, color='b', linewidth=2.5, linestyle='dotted', marker='D', label='y2')
plt.plot(t, t**3, color='g', linewidth=2, linestyle='-.', marker='^', label='y3')

# 設定 圖的標題
plt.title("Line chart")

# 設定 x軸單位
plt.xlabel("x axis")

# 設定 y軸單位
plt.ylabel("y axis")

# 設定 圖例(位置及文字大小)
plt.legend(loc=0, prop={'size': "large"})
```

### 散布圖的格式

`plt.scatter(x, y)`

```python
# 設定 point_x 與 point_y
point_x = [1,2,5,6,3,4,9,8]
point_y = [3,1,8,2,4,3,2,7]

plt.scatter(point_x, point_y)
```

```python
# 設定 point_x 與 point_y
point_x = [1,2,5,6,3,4,9,8]
point_y = [3,1,8,2,4,3,2,7]

# 隨機生成顏色
color = np.random.rand(8)

# 隨機設定點的面積(放大30倍)
area = np.pi * (30 * np.random.rand(8)) ** 2

# s:面積；c:顏色；alpha:透明度(越小越透明)
plt.scatter(point_x, point_y, s=area, c=color, alpha=0.5)
```

### 圓餅圖 : 多種類比例的數據

`plt.pie(x, labels=label, colors=color)`

* `x` : label的量值，例: `abnormal_num`  
* `labels` : 每個種類的標籤  
* `colors` : 每個種類的顏色  

```python
# 設定四個異常種類的數量
abnormal_num = [3, 7, 5, 10]

# 設定四個種類的標籤
label = ["abnormal_1", "abnormal_2", "abnormal_3", "abnormal_4"]

# 設定四個種類的顏色
color = ["red", "yellow", "blue", "green"]

plt.pie(abnormal_num, labels=label, colors=color)
```

### 創建子圖

`fig.add_subplot(row_len, col_len, num)`

* row_len : 第幾列
* col_len : 第幾行
* num : 第幾張圖

```python
# 建立背景畫布面板
fig = plt.figure()

# 建立 2 rows 1 column的兩張子圖 : sub1, sub2

sub1 = fig.add_subplot(2, 1, 1) # fig.add_subplot(211)
sub2 = fig.add_subplot(2, 1, 2) # fig.add_subplot(212)

# 在子圖上畫上折線圖
sub1.plot([8,9,5,6,2])
sub2.plot([2,1,9,3,4])
```

### 時序圖 (實例)

* x軸 : 時間
    * `import matplotlib.dates as mdates` : X軸時間設定
* y軸 : 數值

```python
import pandas as pd

# event.csv裡的資料
df = pd.read_csv('event.csv')
df.head(15)
```

* `header` : 指定行數(Row)用來作為列名，數據開始行數(Row)。
* `index_col` : 指定index欄位(可以多行)
* `parse_dates` : True:解析日期index ; False:不解析 
* `squeeze` : 欄位只有一列，則返回一個Series

```python
import matplotlib.dates as mdates

# 將資料的時間欄位設為 index
data = pd.read_csv('event.csv', header=0, index_col=0, parse_dates=True, squeeze=False)
data.head(15)
```

```python
# data = pd.read_csv('event.csv', header=1, index_col=0, parse_dates=True, squeeze=False)
# data.head(15)
```

```python
# 建立fig:畫布面板跟 axes:一張圖表
fig, axes = plt.subplots(figsize=(10,10))
# fig, axes = plt.subplots(2, 2, figsize=(10,10))

# 圖表外觀設定 : seaborn-darkgrid
plt.style.use('seaborn-darkgrid')

# x軸:時間數列 ; y軸:溫度點位的量值 ; 條線的名稱:temperature ; linewidth:2
plt.plot(data.index, data['temperature'], label = 'temperature', linewidth = 2)

# X軸旋轉字體(90度)
plt.xticks(rotation=90)

# WeekdayLocator(星期), DayLocator(日) 
axes.xaxis.set_major_locator(mdates.DayLocator(interval=1))

# 設定時間格式
axes.xaxis.set_major_formatter(mdates.DateFormatter('%Y/%m/%d'))

# x軸單位:Time ; y軸單位:溫度量值 ; 圖的標題:Temperature ; 設定圖示說明:loc=2(左上角)
plt.xlabel("Time(m)")
plt.ylabel("Value")
plt.title("Temperature")
plt.legend(loc=2)

# 儲存成圖片，dpi為解析度
plt.savefig("Temperature.png", dpi = 100)
```
