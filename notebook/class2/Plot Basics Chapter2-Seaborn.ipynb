### 背景style

* darkgrid
* dark
* whitegrid
* white
* ticks

```python
# 設定一個 list point_x 與 list point_y
point_x = [1,2,3,4,5]
point_y = [25,16,9,4,1]

# 設定背景格式 例: darkgrid 灰色網格
sns.set(style = "darkgrid")

plt.plot(point_x, point_y)
```

### set_context() : 用於圖形大小的設置

* paper
* notebook(預設)
* talk
* poster

```python
# 設定 point_x 與 point_y
point_x = np.arange(-180, 180)
point_y = np.sin(point_x * np.pi / 180.0)

sns.set_context('poster')
plt.plot(point_x, point_y)
```

```python
# 設定 point_x 與 point_y
point_x = np.arange(-180, 180)
point_y = np.sin(point_x * np.pi / 180.0)

sns.set_context('talk')
plt.plot(point_x, point_y)
```

### 調色盤

* 顏色 : 
    * `color_palette(palette=None, n_colors=None, desat=None)`
        * palette : 調色盤
            * deep : 深色系列
            * muted : 柔和系列
            * pastel : 粉彩系列
            * bright : 亮色系列
            * dark : 黑暗系列
            * colorblind : 色盲系列
        * n_colors : n個顏色
        * deset : 飽和度，值界在0到1之間  
    * `hls_palette(h = None, l = None, s = None)`
        * h : 色相，意旨色調變化程度，0~1
        * l : 亮度，意旨明度，0~1
        * s : 飽和度，意旨透明度，0~1

```python
# 設定四個異常種類的數量
abnormal_num = [3, 7, 5, 10]

# 設定四個種類的標籤
label = ["abnormal_1", "abnormal_2", "abnormal_3", "abnormal_4"]

# 設定四個種類的顏色
color = ["red", "yellow", "blue", "green"]

plt.pie(abnormal_num, labels=label, colors=color)
```

`sns.color_palette("muted", n_colors, desat)`

* `'muted'` : 柔和模式 
* `n_color` : 設定要幾種顏色
* `desat` : 飽和度

```python
# 設定調色盤 : 柔和模式
muted_palette = sns.color_palette("muted", n_colors=15, desat=0.5)

# 顯示調色盤
sns.palplot(muted_palette)
```

```python
# 設定四個異常種類的數量
abnormal_num = [3, 7, 5, 10]

# 設定四個種類的標籤
label = ["abnormal_1", "abnormal_2", "abnormal_3", "abnormal_4"]

# colors : 柔和模式
plt.pie(abnormal_num, labels=label, colors=muted_palette)
```

`sns.hls_palette(h, l, s)`

* h : 色相 (0~1)
* l : 亮度 (0~1)
* s : 飽和度 (0~1)

```python
# 設定調色盤 色相、亮度、飽和度
hls_palette = sns.hls_palette(h = 0.3, l = 0.8, s = 0.7)

# 顯示調色盤
sns.palplot(hls_palette)
```

```python
# 設定四個異常種類的數量
abnormal_num = [3, 7, 5, 10]

# 設定四個種類的標籤
label = ["abnormal_1", "abnormal_2", "abnormal_3", "abnormal_4"]

# colors : hls_palette
plt.pie(abnormal_num, labels=label, colors=hls_palette)
```
